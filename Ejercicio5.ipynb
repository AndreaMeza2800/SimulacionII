#Ejercicio 5:
a) Calcular E(Zn) y V(Zn) 
b) ¿Cuál debe ser el valor de N para que el error sea de 0.01?

```python
import random
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)
y = np.sqrt(1 - x**2)               # Ecuación del círculo en el primer cuadrante 
fig = plt.figure(1, figsize = (10, 10)) 
plt.plot(x, y, color = 'blue', markersize = 1) 
def pi_mc():
    pi = 0
    counter = 0 # numero de dardos dentro 
    xi=[] #Lista de xi
    e=10
    n=1
    while e >.01:
        x0 = random.random()
        y0 = random.random()
        plt.scatter(x0, y0, s = 20, c = 'red')  
        if x0**2 + y0**2 <1:
          counter = counter + 1
          xi.append(1)# xi toma valor de uno (dentro del circulo)
        else:
          xi.append(0)# xi toma valor de cero (fuera del circulo)
        pi = 4.0*float(counter)/(n + 1)
        n=n+1
        e = np.absolute((pi - np.pi)/np.pi)
    Z=sum(xi)/n
    V=(Z)*(1-(Z))
    print("pi =", pi)
    print("error =", e)
    print("a) Calcular E(Zn) y V(Zn)")
    # E(Zn)=p =Z
    print("E(x_i)=",Z)
    # V(Zn)=pq = Z(1-Z)
    print("V(x_i)=",V)
    print("¿Cuanto vale N para que el error sea <= a .01?")
    print("N=",n)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid()
    plt.show()
pi_mc()
```
