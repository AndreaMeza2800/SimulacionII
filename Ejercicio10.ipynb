Generar muestras e histograma de las funciones: triangular(), betavariate(), expovariate(), gammavariate(), lognormvariate(), normalvariate(), vonmisesvariate(), paretovariate() y weibullvariate()

```python
import random
import numpy as np
import matplotlib.pyplot as plt
#Generamos una lista de 100 n√∫meros aleatorios normalmente distribuidos.
nums = []
mu = 100
sigma = 50

for i in range(200):
    temp = random.gauss(mu, sigma)
    nums.append(temp)
plt.plot(nums)
plt.show()
# Construimos el histograma de una muestra de 10000 n√∫meros aleatorios normalmente distribuidos
for i in range(10000):
    temp = random.gauss(mu, sigma)
    nums.append(temp)
plt.hist(nums, bins = 200)
plt.show()
```

```python
#Funcion triangular
import random
import numpy as np
import matplotlib.pyplot as plt
#Generamos una lista de 100 n√∫meros aleatorios dentro de un rango con un sesgo hacia un extremo
nums = []
for i in range(200):
    temp = random.triangular(0,100,20) #(Limite inferior ,limite superior, moda)
    nums.append(temp)
plt.plot(nums)
plt.show()
# Construimos el histograma de una muestra de 10000 n√∫meros aleatorios con random triangular
for i in range(10000):
    temp = random.triangular(0,100,20)
    nums.append(temp)
plt.hist(nums, bins = 200)
plt.show()
```

```python
#Funcion betavariate
import random
import numpy as np
import matplotlib.pyplot as plt
#Generamos una lista de 100 n√∫meros aleatorios con distribuci√≥n beta . El valor devuelto est√° entre 0 y 1.
nums = []
for i in range(200):
    temp = random.betavariate(5,10) #(alpha,beta) alpha,beta >0
    nums.append(temp)
plt.plot(nums)
plt.show()
# Construimos el histograma de una muestra de 10000 n√∫meros aleatorios con random.betavariate
for i in range(10000):
    temp = random.betavariate(5,10)
    nums.append(temp)
plt.hist(nums, bins = 200)
plt.show()
```

```python
#Funcion expovariable
import random
import numpy as np
import matplotlib.pyplot as plt
#Generamos una lista de 100 n√∫meros aleatorios con distribuci√≥n exponencial.si el par√°metro es positivo, los resultados van de 0 a infinito positivo ,si el par√°metro es negativo, los resultados van de 0 a infinito negativo
nums = []
for i in range(200):
    temp = random.expovariate(2) #(lambda) lambda diferente de 0
    nums.append(temp)
plt.plot(nums)
plt.show()
# Construimos el histograma de una muestra de 10000 n√∫meros aleatorios con random.expovariable
for i in range(10000):
    temp = random.expovariate(2)
    nums.append(temp)
plt.hist(nums, bins = 200)
plt.show()
```

```python
#Funcion gammavariate
import random
import numpy as np
import matplotlib.pyplot as plt
#Generamos una lista de 100 n√∫meros aleatorios con distribuci√≥n gamma
nums = []
for i in range(200):
    temp = random.gammavariate(5,10) #(alpha,beta) alpha,beta >0
    nums.append(temp)
plt.plot(nums)
plt.show()
# Construimos el histograma de una muestra de 10000 n√∫meros aleatorios con random.betavariate
for i in range(10000):
    temp = random.gammavariate(5,10)
    nums.append(temp)
plt.hist(nums, bins = 200)
plt.show()
```

```python
#Funcion lognormvariate
import random
import numpy as np
import matplotlib.pyplot as plt
#Generamos una lista de 100 n√∫meros aleatorios con distribuci√≥n logaritmica normal 
nums = []
mu = 0
sigma = .25
for i in range(200):
    temp = random.lognormvariate(mu, sigma)
    nums.append(temp)
plt.plot(nums)
plt.show()
# Construimos el histograma de una muestra de 10000 n√∫meros aleatorios con lognormvariate
for i in range(10000):
    temp = random.lognormvariate(mu, sigma)
    nums.append(temp)
plt.hist(nums, bins = 200)
plt.show()
```

```python
#Funcion  normalvariate
import random
import numpy as np
import matplotlib.pyplot as plt
#Generamos una lista de 100 n√∫meros aleatorios con distribuci√≥n  normal 
nums = []
mu = 100
sigma = 50
for i in range(200):
    temp = random.normalvariate(mu,sigma)
    nums.append(temp)
plt.plot(nums)
plt.show()
# Construimos el histograma de una muestra de 10000 n√∫meros aleatorios con normalvariate
for i in range(10000):
    temp = random.normalvariate(mu, sigma)
    nums.append(temp)
plt.hist(nums, bins = 200)
plt.show()
```

```python
#Funcion  vonmisesvariate
import random
import numpy as np
import matplotlib.pyplot as plt
#Generamos una lista de 100 n√∫meros aleatorios con distribuci√≥n de von Mises o distribuci√≥n normal circular .
nums = []
mu=0 #√°ngulo medio, expresado en radianes entre 0 y 2*pi
kappa=4 # par√°metro de concentraci√≥n, mayor o igual a cero
for i in range(200):
    temp = random.vonmisesvariate(mu,kappa)
    nums.append(temp)
plt.plot(nums)
plt.show()
# Construimos el histograma de una muestra de 10000 n√∫meros aleatorios con vonmisesvariate
for i in range(10000):
    temp = random.vonmisesvariate(mu,kappa)
    nums.append(temp)
plt.hist(nums, bins = 200)
plt.show()
```

```python
#Funcion  paretovariate 
import random
import numpy as np
import matplotlib.pyplot as plt
#Generamos una lista de 100 n√∫meros aleatorios con distribuci√≥n de Pareto 
nums = []
for i in range(2000):
    temp = random.paretovariate(5) #alpha
    nums.append(temp)
plt.plot(nums)
plt.show()
# Construimos el histograma de una muestra de 10000 n√∫meros aleatorios con paretovariate 
for i in range(10000):
    temp = random.paretovariate(5) 
    nums.append(temp)
plt.hist(nums, bins = 200)
plt.show()
```

```python
#Funcion  weibullvariate
import random
import numpy as np
import matplotlib.pyplot as plt
#Generamos una lista de 100 n√∫meros aleatorios con distribuci√≥n Weibull
nums = []
for i in range(200):
    temp = random.gammavariate(5,1) #(alpha,beta) alfa: par√°metro de escala (.5,1,1.5,5) beta: par√°metro de forma 
    nums.append(temp)
plt.plot(nums)
plt.show()
# Construimos el histograma de una muestra de 10000 n√∫meros aleatorios con weibullvariate
for i in range(10000):
    temp = random.gammavariate(5,1)
    nums.append(temp)
plt.hist(nums, bins = 200)
plt.show()
```
