1. Mostrar que $\theta_1 $ es un estimador insesgado, es decir, que $E(\theta_1 ) = pc(b-a) = I$
2. Mostrar que $V(\theta_1 )=\frac{I}{N}[c(b-a)-I]$

```python
import random                                   # importamos el módulo random
import numpy as np                              # importamos numpy con un alias (np)
import matplotlib.pyplot as plt                 # ídem
import statistics as stats   
import pandas as pd
#Definimos la función de prueba.
def test_function(x):
    return(np.sqrt(np.arctan(x)))
def integrate(x1,x2,func=test_function,n=100000):
    X=np.linspace(x1,x2,1000)
    y1=0
    y2=max((func(X)))
    cc=y2
    area=(x2-x1)*(y2-y1)
    check=[]
    xs=[]
    ys=[]
    sum=0
    for i in range(n):
        x=np.random.uniform(x1,x2)
        xs.append(x)
        y=np.random.uniform(y1,y2)
        ys.append(y)
        if abs(y)>abs(func(x)) or y<0:
            check.append(0)
        else:
            check.append(1)
            sum=sum+1
    p=sum/n
    return(cc,p,np.mean(check)*area,xs,ys,check)
cc,p,I,x,y,c=integrate(0,1)
print("El valor de la integral es:",I)
_,_,_,x,y,c=integrate(0,1,n=500)
df=pd.DataFrame()
df['x']=x
df['y']=y
df['c']=c
b=1
a=0
E=(p*cc)*(b-a) # Caculamos E(theta1)
V=(I/100000)*(cc*(b-a)-I)
print("El valor de E(theta1)=",E)
print("Por lo tanto , E(theta1) = I")
print("V=(I/N)(c(b-a)-I) =",V)
X=np.linspace(0,1,1000)
plt.grid(True)
plt.plot(X,test_function(X))
plt.scatter(df[df['c']==0]['x'],df[df['c']==0]['y'],color='red')
plt.scatter(df[df['c']==1]['x'],df[df['c']==1]['y'],color='blue')
plt.show()
```
