a) Calcular E(x_i) y V(x_i)

```python
import random
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)
y = np.sqrt(1 - x**2)               # Ecuación del círculo en el primer cuadrante 
fig = plt.figure(1, figsize = (10, 10)) 
plt.plot(x, y, color = 'blue', markersize = 1) 
def pi_mc(n):
    pi = 0
    counter = 0 # numero de dardos dentro 
    xi=[] #Lista de xi
    for i in range(n):
        x0 = random.random()
        y0 = random.random()
        plt.scatter(x0, y0, s = 20, c = 'red')  
        if x0**2 + y0**2 <1:
          counter = counter + 1
          xi.append(1)# xi toma valor de uno (dentro del circulo)
        else:
          xi.append(0)# xi toma valor de cero (fuera del circulo)
        pi = 4.0*float(counter)/(i + 1)
        e = np.absolute((pi - np.pi)/np.pi)
    Z=sum(xi)/n
    V=(Z)*(1-(Z))
    print("pi =", pi)
    print("error =", e)
    # E(xi)=p =Z
    print("E(x_i)=",Z)
    # V(xi)=pq = Z(1-Z)
    print("V(x_i)=",V)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid()
    plt.show()
pi_mc(200)
```
