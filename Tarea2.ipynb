*Calcular la integral de raiz cuadrada de arco tangente de 0 a 1, usando un método de MC similar al que usamos para estimar pi*

```python
import numpy as np 
import math
import random
import matplotlib.pyplot as plt 
x = np.linspace(0,1,100)
y = np.sqrt(np.arctan(x)) # Curva
contador = 0 
n = 80 # de dardos
d = 0 #dardos dentro de la circunferencia
fig = plt.figure(1, figsize = (10, 10)) 
plt.plot(x, y, color = 'blue', markersize = 1) 
for i in range(n):
  x0 = random.random()
  y0 = random.random()
  h = ((math.atan(x0)**(1/2))/y0)
  plt.scatter(x0, y0, s = 20, c = 'red')  
  if h < 1:#dardo dentro de la función
    d = d+1 
  contador = contador+1 
tan = 4*(d/n)
print ("Raíz del arco tangente de 0 a 1 es: ",tan)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
