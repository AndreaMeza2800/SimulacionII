Estimar pi utilizando un cuarto de elipse y visualizar el comportamiento de esta estimación en función del número de iteraciones

```python
#Andrea Meza 
import random
import numpy as np
import matplotlib.pyplot as plt

N = 1000 # Numero de dardos
Aciertos = 0
b=5
a=3.5

XElipse=[]  
YElipse=[]  
XSquare=[]  
YSquare=[]  

for p in range(N):
    x=random.uniform(0,a) #Devuelve un numero flotante aleatorio entre 0 y a
    y=random.uniform(0,b) #Devuelve un numero flotante aleatorio entre 0 y b
    if(((x/a)**2+(y/b)**2) <= 1):
        Aciertos+=1 #Contador de dardos dentro del cuarto de elipse
        XElipse.append(x)  
        YElipse.append(y)        
    else:
        XSquare.append(x)  
        YSquare.append(y)

Pi = 4*Aciertos/N #Aproximacion de pi

print("N=%d Aciertos=%d Pi=%.2f" %(N,Aciertos,Pi))

XLin=np.linspace(0,a)
YLin=[]
for x in XLin:
    YLin.append(np.sqrt((1-(x/a)**2)*b*b)) #Ecuacion de elipse

fig = plt.figure(1, figsize=(8, 8)) 
plt.axis("equal")                             
plt.grid(which="major")                        
plt.plot(XLin , YLin, color="red" , linewidth="1")  #Grafica Elipse
plt.scatter(XElipse, YElipse, color="blue", marker   =".")  #Grafica dardos dentro de cuarto de elipse
plt.scatter(XSquare, YSquare, color="green"  , marker   =".")  #Grafica dardos fuera de cuarto de elipse
plt.title("Estimación de $\pi$ por el método de Monte Carlo")
plt.xlabel("X")
plt.ylabel("Y")
 
plt.show()
```

```python
def pi(n):
    pi = 0
    lx = []
    ly = []
    b=5
    a=3.5
    counter = 0
    for i in range(n):
        x=random.uniform(0,a) #Devuelve un numero flotante aleatorio entre 0 y a
        y=random.uniform(0,b) #Devuelve un numero flotante aleatorio entre 0 y b
        if (((x/a)**2+(y/b)**2) <= 1):
            counter = counter + 1
            lx.append(counter)
            ly.append(4.0*float(counter)/(i + 1))
        
        pi = 4.0*float(counter)/(i + 1)
        
        e = np.absolute((pi - np.pi)/np.pi) #Calculo de error absoluto
    print("pi =", pi)
    print("error =", e)
    
    fig = plt.figure(1, figsize=(8, 8))        
    plt.xlabel("Iteraciones")
    plt.ylabel("Estimación de $\pi$")
    plt.grid()
    plt.plot(lx, ly, color="blue", markersize=1)  #comportamiento de estimación  dee pi en función del número de iteraciones.
    plt.show()
pi(1000)
```
